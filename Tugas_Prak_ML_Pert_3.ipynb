## Modul 2

```python
import pandas as pd

# Load data
data = pd.read_csv("DatasetForCoffeeSales2.csv")
display(data.head(10))

# Filter data (City, Category, Product, Quantity)
data.filter(["City", "Category", "Product", "Quantity"]).head()

# Lakukan Sorting Quantity (most)
data.sort_values("Quantity", axis=0, ascending=False, inplace=True, na_position="last")
display(data.head())

# Grouping untuk melihat total sales pada setiap City
grp1 = data.groupby('City')
result = grp1['Final Sales'].aggregate('sum')
print(result)
```

## Modul 3

```python
import pandas as pd

# Load data
data = pd.read_csv("students.csv")
display(data.head(10))

# Tampilkan informasi mengenai setiap fitur yang dimiliki menggunakan info()
data.info()

# Gunakan fungsi isna() dan sum() pada dataset untuk missing value handling
data.isna().sum()

# Handling missing value menggunakan nilai modus-nya
data['lunch'].fillna(data['lunch'].mode()[0], inplace=True)

# Handling missing value menggunakan nilai mean (karena tipe data fitur reading score adalah float)
data['reading score'].fillna(data['reading score'].mean(), inplace=True)

# Handling missing value menggunakan nilai median (karena tipe data fitur reading score adalah float dan bertipe ordinal)
data['grade'].fillna(data['grade'].median()[0], inplace=True)

# Cek kembali apakah masih ada nilai Nan pada data menggunakan fungsi info()
data.info()

# Interpolate the Missing Value
data['nama_fitur'].interpolate(methode='linear', inplace=True)

#backwardfill the missing value
data['nama_fitur'].filna(method= 'bfill', inplace = True)

#forwardfill the missing value
data['nama_fitur'].filna(method= 'bfill', inplace = True)

# drop Nan Value (Opsi 1)
data['nama_fitur'].dropna(axis=0, inplace=True)

# drop feature with more than 50% are Nan value (Opsi 2)
data['nama_fitur'].drop(axis=1, inplace=True)
```
